```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
datafram = pd.read_csv("train.csv")
datatest = pd.read_csv("test.csv")
```

```python
# train data
datafram.head()
```

```python
#test data
datatest.head()
```

```python
pd.isnull(datatest).sum()
```

```python
datacombine = [datafram, datatest]
```

```python
pd.isnull(datafram).sum()
```

```python
for dataset in datacombine: 
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)
```

```python
pd.isnull(datatest).sum()
```

```python
datafram.head()
```

```python
pd.isnull(datafram).sum()
```

```python
for dataset in datacombine:    
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
    dataset['IsAlone'] = 1 
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0
```

```python
pd.isnull(datafram).sum()
```

```python
datafram.head()
```

```python
datafram['FareBin'] = pd.qcut(datafram['Fare'], 4)
datatest['FareBin'] = pd.qcut(datatest['Fare'], 4)
datafram.head()
```

```python
pd.isnull(datafram).sum()
```

```python
datafram['AgeBin'] = pd.cut(datafram['Age'].astype(int), 5)
datatest['AgeBin'] = pd.cut(datatest['Age'].astype(int), 5)
datafram.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
datafram['Sex']= label_encoder.fit_transform(datafram['Sex']) 
datatest['Sex']= label_encoder.fit_transform(datatest['Sex']) 

for dataset in datacombine:    
    dataset['AgeBin_Code'] = label_encoder.fit_transform(dataset['AgeBin'])
    dataset['FareBin_Code'] = label_encoder.fit_transform(dataset['FareBin'])
```

```python
pd.isnull(datafram).sum()
```

```python
datafram['Embarked'].fillna("S",inplace=True)
```

```python
datafram['Embarked'].value_counts()
```

```python
embarked_mapping = {"S": 1, "C": 2, "Q": 3}
datafram['Embarked'] = datafram['Embarked'].map(embarked_mapping)
datatest['Embarked'] = datatest['Embarked'].map(embarked_mapping)
```

```python
datatest.head()
```

```python
#แปลงอายุให้เป็นช่วงคน
bins = [-1, 0, 5, 12, 18, 24, 35, 60, np.inf]
labels = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Senior']
datafram['AgeGroup'] = pd.cut(datafram["Age"], bins, labels = labels)
datatest['AgeGroup'] = pd.cut(datatest["Age"], bins, labels = labels)
```

```python
# ตัดเอา คำนำหน้าออก
for dataset in datacombine:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
for dataset in datacombine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Royal": 5, "Rare": 6}
for dataset in datacombine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
datafram.head()
```

```python
age_title_mapping = {1: "Young Adult", 2: "Student", 3: "Adult", 4: "Baby", 5: "Adult", 6: "Adult"}
for x in range(len(datafram["AgeGroup"])):
    if datafram["AgeGroup"][x] == "Unknown":
        datafram["AgeGroup"][x] = age_title_mapping[datafram["Title"][x]]
        
for x in range(len(datatest["AgeGroup"])):
    if datatest["AgeGroup"][x] == "Unknown":
        datatest["AgeGroup"][x] = age_title_mapping[datatest["Title"][x]]
```

```python
#แปลงช่วงอายุเป็นตัวเลข
age_mapping = {'Baby': 1, 'Child': 2, 'Teenager': 3, 'Student': 4, 'Young Adult': 5, 'Adult': 6, 'Senior': 7}
datafram['AgeGroup'] = datafram['AgeGroup'].map(age_mapping)
datatest['AgeGroup'] = datatest['AgeGroup'].map(age_mapping)
```

```python
datafram.head()
```

```python
for x in range(len(datatest["Fare"])):
    if pd.isnull(datatest["Fare"][x]):
        pclass = datatest["Pclass"][x] #Pclass = 3
        datatest["Fare"][x] = round(datafram[datafram["Pclass"] == pclass]["Fare"].mean(), 4)
```

```python
datafram['FareBand'] = pd.qcut(datafram['Fare'], 4, labels = [1, 2, 3, 4])
datatest['FareBand'] = pd.qcut(datatest['Fare'], 4, labels = [1, 2, 3, 4])
```

```python
datafram = datafram.drop(['Age'], axis = 1)
datatest = datatest.drop(['Age'], axis = 1)
```

```python
datafram["CabinBool"] = (datafram["Cabin"].notnull().astype('int'))
datatest["CabinBool"] = (datatest["Cabin"].notnull().astype('int'))
```

```python
pd.isnull(datatest).sum()
```

```python
datafram.drop('Ticket',axis=1,inplace=True)
datafram.drop('Cabin',axis=1,inplace=True)
datafram.drop('Name',axis=1,inplace=True)

datatest.drop('Ticket',axis=1,inplace=True)
datatest.drop('Cabin',axis=1,inplace=True)
datatest.drop('Name',axis=1,inplace=True)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forest
param_grid = { "criterion" : ["gini", "entropy"], "min_samples_leaf" : [1, 5, 10, 25, 50, 70], "min_samples_split" : [2, 4, 10, 12, 16, 18, 25, 35], "n_estimators": [100, 400, 700, 1000, 1500]}
rf = RandomForestClassifier(n_estimators=100, max_features='auto', oob_score=True, random_state=0, n_jobs=-1)
clf = GridSearchCV(estimator=rf, param_grid=param_grid, n_jobs=-1)
clf.fit(x_train, y_train)
```

```python
ids = datatest['PassengerId']
predictions = clf.predict(datatest.drop('PassengerId', axis=1))
```
